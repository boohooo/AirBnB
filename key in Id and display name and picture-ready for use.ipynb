```python
# Basic Libraries
import numpy as np
import pandas as pd
import ipywidgets as widgets #widgets to make fancy interactive elements
import datetime
from IPython.display import IFrame

from PIL import Image #Python Imaging
import requests #Requests is a HTTP library for Python
from io import BytesIO
```

```python
from IPython.display import Markdown, display

def printmd(string): #type printmd() instead of print() and add ** before after string (before the quotation marks) to make the string bold
    display(Markdown(string))
```

```python
listings_data = pd.read_csv('listings.csv')
```

```python
print('Key in your desired listing id')
userinput =  input() #input data type is a string
```

```python
url = listings_data.loc[listings_data.id == int(userinput)]['picture_url']
selectedrow = listings_data.loc[listings_data.id == int(userinput)]
```

```python
name = selectedrow['name'].to_string(index=False) #print selected row without index
print(name, end='') #print without quote thing at the side.

#get image from url, open image.
response = requests.get(selectedrow.picture_url.iloc[0]) #Use iloc to access by position (rather than label(aka index))
img = Image.open(BytesIO(response.content))
display(img)
#img.show() #use .show() when u want it to open outside
```
